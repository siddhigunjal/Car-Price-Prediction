```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Clean_Dataset.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a_cat=df.select_dtypes(object)
a_num=df.select_dtypes(['int','float'])
```

```python
a_num
```

```python
a_num.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
a_num
```

```python
a_cat
```

```python
a_cat.drop('flight',axis=1,inplace=True)
```

```python
a_cat
```

```python
for i in a_cat:
    le=LabelEncoder()
    a_cat[i]=le.fit_transform(a_cat[i])
```

```python
a_cat
```

```python
new=pd.concat([a_cat,a_num],axis=1)
```

```python
new
```

```python
new.info()
```

```python
x=new.iloc[:,:-1]
```

```python
x
```

```python
y=new['price']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
LR=LinearRegression()
LR.fit(xtrain,ytrain)
ypred=LR.predict(xtest)
```

```python
r2_score(ytest,ypred)
```

```python
LR.score(xtrain,ytrain)
```

```python
LR.score(xtest,ytest)
```

```python
#here both training error and testing is low so this not overfitting not underfitting pbm hence it is perfect fit 
#no need to regularisatio
```
